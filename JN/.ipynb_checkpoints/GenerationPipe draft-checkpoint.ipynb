```python
textText1 = r'בהמשך השנה מונה לסמ"ט ב\' בטייסת הנגב. ב־20 בינואר 1982 התנגשו שני מטוסי F-16A ישראליים מהטייסת (מספרי זנב: 276 ו־266), באחד מהמטוסים היה סרן אילן רמון, שהצליח לנטוש בהצלחה. שני המטוסים התרסקו. שנה לאחר מכן השתתף במלחמת לבנון הראשונה.'
textText1 = r'טקסט F-16 טקסט ADFGDASFG טקסט ADSFG-ASDF. A.'

punctList = [r'.', r',', r'?',r'!',r'(',r')', r'—', r'–' ,r'\-', r'־' , r'"', r':', r';', '\'', r'\[', r'\]']
punct = r''.join(punctList)
print(punct)
pattern = re.compile(r'(?:[\s' + punct +'])?([a-zA-Z0-9]+[' + punct + r']*[a-zA-Z0-9]+)(?:[\s' + punct +']?)')
print(r' [a-zA-Z' +  puct + r']+ ')

def replaceEnglish(inText, punctList, englishString = '<אנגלית>'):
    # function replaces 
    punct = r''.join(punctList)
    pattern = re.compile(r'[a-zA-Z0-9]+[' + punct + r']*[a-zA-Z0-9]+')
    outText = pattern.sub(englishString, inText)
    return outText

out = replaceEnglish(textText1, punctList, '<אנגלית>')
print(out)
```
